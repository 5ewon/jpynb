```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumBarunGothic'
from google.colab import files

from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd

csv_file_path = "/content/drive/MyDrive/older.csv"

data = pd.read_csv(csv_file_path, encoding="cp949")
```

```python
data.head()
```

```python
value_counts = data["관리기관명"].value_counts()


plt.figure(figsize=(12, 6))
value_counts.plot(kind='bar')
plt.xlabel("관리기관명")
plt.ylabel("빈도수")
plt.title("senior citizen center")
plt.show()
```

```python
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째 위치
value_counts.plot(kind='pie', autopct='%1.1f%%')  # 원 그래프 생성
plt.ylabel("")  # y축 레이블 설정 (비워두기)
plt.title("senior citizen center")  # 그래프 제목 설정

plt.tight_layout()  # 그래프 간격 조정
```
